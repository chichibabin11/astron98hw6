```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
arr1 = np.random.randint(0,101,40)
arr2 = np.random.randint(0,101,40)
arr3 = np.random.randint(0,101,40)

#vertically 2 stacked
fig, axs = plt.subplots(2,1)

axs[0].plot(arr1, color = 'orange', linewidth = 10)
axs[0].plot(arr2, color = 'red', linestyle = 'dashed')

axs[1].scatter(range(len(arr3)),arr3, color = 'magenta', marker = 'D')
plt.xlabel('position in list')
plt.show()
```
