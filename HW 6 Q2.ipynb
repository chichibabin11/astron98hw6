```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Constants
R = 0.083144
a = 16.02
b = 0.1124
n = 1

Pressure = np.linspace(1, 10, 1000)
Volume = np.linspace(10, 30, 1000)
P_grid, V_grid = np.meshgrid(Pressure, Volume)
 
"""
formula:
(P + a*n**2/V**2)*(V-nb)= nRT
rearranging for T:
T = ((P + a*n**2/V**2)*(V-nb))/nR
"""

Temperature = ((P_grid + a*n**2/V_grid**2)*(V_grid - n*b))/(n*R)

plt.figure()
plt.pcolormesh(P_grid, V_grid, Temperature, cmap = 'inferno', vmin = 0, vmax = 500)
plt.colorbar ( label = 'Temperature (K)')
plt.xlabel('Pressure(Bar)')
plt.ylabel('Volume(Litre)')
plt.title('Van der Waals model (Temperature of 1 mole of acetone)')
plt.grid(True)
plt.show()
```
