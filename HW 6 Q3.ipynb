```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def estimate_pi(N):
    points = np.random.rand(int(N), 2)
    #distance = sqrt(x^2+y^2) so if our random points is in the form of (x,y) then x = 0 and y = 1
    distance = np.sqrt(points[:,0]**2 + points[:,1]**2)
    #points inside circle
    distance_in = distance[distance <= 1]
    #pi approx. = 4 * S(circle)/S(square)
    pi_approx = 4 * len(distance_in)/N
    return pi_approx

N_val= [1e3,1e4,1e5,1e6]
pi = [estimate_pi(N) for N in N_val]

for N, pi_val in zip(N_val, pi):
    print(f"N = {N} : π ≈ {pi_val}")
    
```

```python
#for N = 1e4 
import numpy as np
import matplotlib.pyplot as plt

n = 1e4
point = np.random.rand(int(n), 2)
distances = np.sqrt(point[:,0]**2 + point[:,1]**2)
in_circle = distances[distances <= 1]
pi_val = 4 * len(in_circle)/ n
point_in = point[distances <= 1]
point_out = point[distances > 1]

plt.figure()
plt.scatter(point_in[:,0], point_in[:,1], color = 'moccasin')
plt.scatter(point_out[:,0], point_out[:,1], color = 'maroon')
plt.xlim(0,1)
plt.ylim(0,1)
plt.gca().set_aspect('equal', adjustable = 'box')
plt.text(0.2, 0.4, f'π ≈ {pi_val}', fontsize = 12, bbox = dict(facecolor = 'white', alpha = 0.7))
plt.grid(True)
plt.show()
```
